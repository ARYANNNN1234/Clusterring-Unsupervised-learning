```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
x,y_true=make_blobs(n_samples=1000,centers=3,cluster_std=0.6,random_state=42)
```

```python
x
```

```python
df=pd.DataFrame(x,columns=["Feature 1","Feature 2"])
```

```python
df
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(df)
```

```python
x_scaled
```

```python
inertia=[]
k_range=range(1,11)
```

```python
for k in k_range:
    kmeans=KMeans(n_clusters=k,random_state=42)
    kmeans.fit(x_scaled)
    inertia.append(kmeans.inertia_)
```

```python
inertia
```

```python
plt.plot(k_range,inertia,marker='*',color='red')
```

```python
kmeans_final=KMeans(n_clusters=10,random_state=42)
cluster_label=kmeans_final.fit_predict(x_scaled)
cluster_label
```

```python
df['clusters']=cluster_label
```

```python
sns.scatterplot(x=df['Feature 1'],y=df['Feature 2'],hue=df['clusters'],palette='viridis')
```

```python
x,y=make_blobs(n_samples=500,n_features=5,centers=3,cluster_std=1.5,random_state=42)
```

```python
x
```

```python
from sklearn.decomposition import PCA
```

```python
x_scaled=scaler.fit_transform(x)
```

```python
x_scaled
```

```python
pca=PCA(n_components=2)
x_pca=pca.fit_transform(x_scaled)
```

```python
x_pca
```

```python
df_pca=pd.DataFrame(x_pca,columns=['PC1','PC2'])
df_pca['label']=y
```

```python
df_pca
```

```python
sns.scatterplot(x=df_pca['PC1'],y=df_pca['PC2'],hue=df_pca['label'])
```
